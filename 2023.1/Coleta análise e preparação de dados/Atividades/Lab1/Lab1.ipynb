```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import shutil
import datetime
import time
```

```python
#FUNÇÃO PARA BAIXAR IMAGENS
def download_poster(imgsrc,name,caminho):
    response = requests.get(imgsrc, stream=True)
    with open(f'.\{caminho}\\'+name+'.png', 'wb') as f:
        shutil.copyfileobj(response.raw, f)
        return f'{name}.png'
```

```python
#TAREFA 1.3

#LISTA ONDE TEM AS LISTAS DAS INFORMAÇÕES DOS PAÍSES
lista_geral = []
responseT1_1=requests.get(url=r'http://127.0.0.1:8000/places/default/sitemap.xml')
soupT1_1=BeautifulSoup(responseT1_1.content.decode(),'html5lib')
for i in soupT1_1.find_all('loc'):
    tDay = datetime.datetime.now()
    #CAPTURANDO O HTML E ESCREVENDO EM UM ARQUIVO
    nomeArq='Htmls/'+re.findall('.*\/(.*)',i.contents[0])[0]+'.html'
    responsex=requests.get(url=i.contents[0])
    insidesoupT1_1=BeautifulSoup(responsex.content.decode(),'html5lib')
    with open(nomeArq,'w') as file:
        file.write(insidesoupT1_1.prettify())
    #CAPTURANDO O TIMESTAMP
    mes = tDay.month
    dia = tDay.day
    hora = tDay.hour
    minuto = tDay.minute
    horaAtual = f'{dia}/{mes} {hora}:{minuto}'
    #LISTA COM AS INFORMAÇÕES DOS PAÍSES
    lista_pais = []
    urlPais = requests.get(url=i.text)
    soup = BeautifulSoup(urlPais.content.decode(),'html5lib')
    j=soup.find('img')
    name = re.findall('.*\/(.*)',i.contents[0])[0]
    source= r'http://127.0.0.1:8000'+str(j['src'])
    img = download_poster(source,name,'Bandeiras')   
    for i in soup.find_all(class_ = 'w2p_fw'):
        lista_pais.append(i.text)
    lista_pais[0]=img
    lista_pais.append(horaAtual)
    lista_geral.append(lista_pais)    
columns = ['National Flag', 'Area','Population','Iso','Country','Capital','Continent','Tld','Currency Code','Currency Name','Phone','Postal Code Format','Postal Code Regex','Languages','Neighbours','TimeStamp']
df = pd.DataFrame(lista_geral,columns=columns)
```

```python
#TAREFA 1.3

def AtualizaDf(novoDf,velhoDf):
    tDay = datetime.datetime.now()
    dfTeste = velhoDf.drop('TimeStamp',axis=1)
    df_diff = dfTeste.compare(novoDf)
    index_diff = df_diff.index.tolist()
    mes = tDay.month
    dia = tDay.day
    hora = tDay.hour
    minuto = tDay.minute
    horaAtual = f'{dia}/{mes} {hora}:{minuto}'
    novoDf['TimeStamp'] = horaAtual
    velhoDf.loc[index_diff] = novoDf.loc[index_diff]
    return velhoDf
dfFinal = AtualizaDf(df2, df)
dfFinal.to_csv('trabalho.csv')
```
